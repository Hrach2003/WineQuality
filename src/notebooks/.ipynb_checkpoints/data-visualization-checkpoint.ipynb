```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv('../data/winequalityN.csv')

print("rows, columns: ", str(df.shape))
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df = df.fillna(df.mean())
df.isna().sum()
```

## One-hot encode wine type

```python
type_fig = px.histogram(df,x='type')
type_fig.show()
```

```python
df['type'] = pd.get_dummies(df['type'])
df['type'].value_counts()
```

```python
df.describe()
```

## Quantity is in range [3, 9]
Let's plot it.

```python
quality_fig = px.histogram(df,x='quality')
print(df['quality'].unique())
quality_fig.show()
px.box(df, x='quality')
```

```python
corr = df.corr()
plt.subplots(figsize=(15,10))
sns.heatmap(corr, xticklabels=corr.columns, yticklabels=corr.columns, annot=True, cmap=sns.diverging_palette(220, 20, as_cmap=True))
```

```python
from sklearn.model_selection import train_test_split

X = df[df.columns.difference(['quality', 'type', 'free sulfur dioxide'])]
y = df['quality']

```

```python
from imblearn.over_sampling import SMOTE
oversample = SMOTE(k_neighbors=4)

X, y = oversample.fit_resample(X, y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report
model2 = RandomForestClassifier(random_state=1)
model2.fit(X_train, y_train)
y_pred2 = model2.predict(X_test)
print(classification_report(y_test, y_pred2))
```

```python
from sklearn.tree import DecisionTreeClassifier
model3 = DecisionTreeClassifier(random_state=1)
model3.fit(X_train, y_train)
y_pred3 = model3.predict(X_test)
print(classification_report(y_test, y_pred2))
```

```python
from sklearn.svm import SVC
model4 = SVC(random_state=1)
model4.fit(X_train, y_train)
y_pred3 = model4.predict(X_test)
print(classification_report(y_test, y_pred2))
```
